```python
nrow = len(data_subcat)
relation = {}
for i in range(1, nrow):
    relation[i] = [] # i is the parent of a list of children


for r in range(nrow):
    item_id, parent_id = data_subcat.iloc[r]['id'], data_subcat.iloc[r]['parent_id']
    #print(item_id, parent_id)
    relation[parent_id].append(item_id)

# find the lowest categories, the ones with empty children list
lowest_categories = []
for key, value in relation.items():
    if len(value) == 0:
        lowest_categories.append(key)

print(lowest_categories)
```
